```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """Sigmoid function."""
    return 1 / (1 + np.exp(-x))

# Generate a range of values from -10 to 10 for input
x = np.linspace(-10, 10, 100)

# Define different weights and biases to illustrate their effects on the sigmoid function
weights = [0.5, 1, 2]  # Different slopes
biases = [-2, 0, 2]    # Different shifts

plt.figure(figsize=(14, 8))

# Plot the sigmoid function for each combination of weights and biases
for w in weights:
    for b in biases:
        y = sigmoid(w * x + b)  # Apply the sigmoid function with given w and b
        plt.plot(x, y, label=f'w={w}, b={b}')

plt.title('Effect of Weights and Biases on the Sigmoid Function')
plt.xlabel('Input value (x)')
plt.ylabel('Sigmoid Output')
plt.legend()
plt.grid(True)
plt.show()
```
