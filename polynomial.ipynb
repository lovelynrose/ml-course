Add Gaussian Noise to Dataset

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of samples
num_samples = 100

# Create an array of 100 points between 0 and 1
x = np.linspace(0, 1, num_samples)

# Compute the sine of 2*pi*x
y = np.sin(2 * np.pi * x)

# Add Gaussian noise to y
# Set a mean and a standard deviation for the Gaussian noise
noise_mean = 0
noise_std_dev = 0.1

# Generate Gaussian noise
noise = np.random.normal(noise_mean, noise_std_dev, y.shape)

# Create a new y dataset with noise added
y_noisy = y + noise

# Combine x and y into a single dataset for the original and noisy data
dataset_original = np.column_stack((x, y))
dataset_noisy = np.column_stack((x, y_noisy))

# Plot the original dataset
plt.scatter(x, y, label='Original Dataset')

# Plot the noisy dataset
plt.scatter(x, y_noisy, color='red', label='Noisy Dataset', alpha=0.6)

# Label the axes
plt.xlabel('x')
plt.ylabel('y')

# Title of the plot
plt.title('Original vs Noisy Data')

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Show the plot on the screen
plt.show()
```

Fit with Polynomial of Degree 0

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming x, y, and y_noisy are already defined as in previous steps

# Fit a polynomial of degree 0 (constant) to the noisy data
# This is effectively calculating the mean of the y_noisy values
constant_fit = np.polyfit(x, y_noisy, 0)

# Create a line based on the constant fit
y_fit = np.full_like(x, constant_fit)

# Plot the original dataset
plt.scatter(x, y, label='Original Data')

# Plot the noisy dataset
plt.scatter(x, y_noisy, color='red', label='Noisy Data', alpha=0.6)

# Plot the line of best fit
plt.plot(x, y_fit, color='green', label='Degree 0 Fit', linewidth=2)

# Label the axes
plt.xlabel('x')
plt.ylabel('y')

# Title of the plot
plt.title('Polynomial Degree 0 Fit to Noisy Data')

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Show the plot on the screen
plt.show()
```

Residuals

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate dataset
np.random.seed(0)  # for reproducibility
x = np.linspace(0, 1, 10)  # 10 data points between 0 and 1
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.1, 10)  # sin(2*pi*x) with noise

# Reshape x for sklearn
X = x.reshape(-1, 1)

# Fit linear regression
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plotting
plt.figure(figsize=(8, 6))

# Plotting the data and linear fit
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Linear Fit')

# Plotting the residuals as lines
for xi, yi, yp in zip(X, y, y_pred):
    plt.plot([xi, xi], [yi, yp], color='green', linestyle='--') # Residual lines

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data, Linear Fit, and Residuals')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import seaborn as sns

# Generate dataset
np.random.seed(0)  # for reproducibility
x = np.linspace(0, 1, 10)  # 10 data points between 0 and 1
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.1, 10)  # sin(2*pi*x) with noise

# Reshape x for sklearn and create polynomial features
X = x.reshape(-1, 1)
poly = PolynomialFeatures(degree=3)  # Using a cubic polynomial
X_poly = poly.fit_transform(X)

# Fit linear regression to the polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Calculate residuals
residuals = y - y_pred

# Plotting
plt.figure(figsize=(12, 6))

# Plot 1: Data and Polynomial Regression Fit
plt.subplot(1, 2, 1)
plt.scatter(X, y, label='Data')
# Plot the polynomial regression fit
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_poly = poly.transform(x_range)
plt.plot(x_range, model.predict(y_range_poly), color='red', label='Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Polynomial Regression Fit')
plt.legend()

# Plot 2: Distribution of Residuals
plt.subplot(1, 2, 2)
sns.histplot(residuals, kde=True, color='green')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import seaborn as sns

# Generate dataset
np.random.seed(0)  # for reproducibility
x = np.linspace(0, 1, 100)  # 10 data points between 0 and 1
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.1, 100)  # sin(2*pi*x) with noise

# Reshape x for sklearn and create polynomial features
X = x.reshape(-1, 1)
poly = PolynomialFeatures(degree=3)  # Using a cubic polynomial
X_poly = poly.fit_transform(X)

# Fit linear regression to the polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Calculate residuals
residuals = y - y_pred

# Plotting
plt.figure(figsize=(12, 6))

# Plot 1: Data and Polynomial Regression Fit
plt.subplot(1, 2, 1)
plt.scatter(X, y, label='Data')
# Plot the polynomial regression fit
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_poly = poly.transform(x_range)
plt.plot(x_range, model.predict(y_range_poly), color='red', label='Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Polynomial Regression Fit')
plt.legend()

# Plot 2: Distribution of Residuals
plt.subplot(1, 2, 2)
sns.histplot(residuals, kde=True, color='green')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')

plt.tight_layout()
plt.show()
```
